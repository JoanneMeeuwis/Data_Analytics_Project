```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/celestinefeuillat/Desktop/training.1600000.processed.noemoticon.csv", nrows = 200000)
```

```python
data.columns
data.columns = ['Target', 'Id','Date', 'Flag', 'User', 'Text' ]
```

```python
data["num_char"] = data['Text'].apply(len)
```

```python
def hashtag_count(string) :
    words = string.split()
    hashtags = [word for word in words if word.startswith('#')]
    return len(hashtags)

def mention_count(string) :
    words = string.split()
    mention = [word for word in words if word.startswith('@')]
    return len(mention)
```

```python
data["num_hashtag"] = data["Text"].apply(hashtag_count)
data["num_mention"] = data["Text"].apply(mention_count)
print(data["num_mention"].mean())
```
